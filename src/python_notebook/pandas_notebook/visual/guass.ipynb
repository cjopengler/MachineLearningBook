# 高斯曲线绘制

```python
import seaborn as sns
from pandas import DataFrame, Series
import pandas as pd

%matplotlib inline
```

```python
titanic_df = pd.read_csv('./data/train.csv')
```

年龄数据是高斯分布的，下面会根据年龄按照 'Survived' 分类来进行高斯曲线绘制。

```python
facet = sns.FacetGrid(titanic_df,
                          hue='Survived',
                          aspect=4)
facet.map(sns.kdeplot, 'Age', shade=True)
facet.set(xlim=(0, titanic_df['Age'].max()))
facet.add_legend()
```

参数解释， aspect 是用来控制显示区域的大小，4表示4倍的意思。'hue' 是分类的标准。col: 是按照行产生分类。现在的这个数据因为 'Servived'是2类，所以应该产生两张图片.

```python
facet = sns.FacetGrid(titanic_df,
                          col='Survived',
                          aspect=4)
facet.map(sns.kdeplot, 'Age', shade=True)
facet.set(xlim=(0, titanic_df['Age'].max()))
facet.add_legend()
```

看到当col设置成'Survived'之后就会按照这种方式来显示图像
